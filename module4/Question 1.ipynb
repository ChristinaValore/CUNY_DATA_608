### Question 1: What proportion of trees are in good, fair, or poor health according to the ‘health’ variable?

```python
import pandas as pd 
import numpy as np
import datetime

import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
```

```python
url = 'https://data.cityofnewyork.us/resource/nwxe-4ae8.json'
trees = pd.read_json(url)
trees.head(5)
```

```python
soql_url = ('https://data.cityofnewyork.us/resource/nwxe-4ae8.json?' +\
        '$select=health,count(tree_id)' +\
        '&$where=boroname=\'Bronx\'' +\
        '&$group=health').replace(' ', '%20')
soql_url_bx = pd.read_json(soql_url)
soql_url_bx
soql_url_bx.dropna()
```

```python
soql_url = ('https://data.cityofnewyork.us/resource/nwxe-4ae8.json?' +\
        '$select=health,count(tree_id)' +\
        '&$where=boroname=\'Brooklyn\'' +\
        '&$group=health').replace(' ', '%20')
soql_url_bk = pd.read_json(soql_url)
soql_url_bk.dropna()
```

```python
soql_url_si = ('https://data.cityofnewyork.us/resource/nwxe-4ae8.json?' +\
        '$select=health,count(tree_id)' +\
        '&$where=boroname=\'Staten Island\'' +\
        '&$group=health').replace(' ', '%20')
soql_url_si = pd.read_json(soql_url)
soql_url_si.dropna()
```

```python
soql_url = ('https://data.cityofnewyork.us/resource/nwxe-4ae8.json?' +\
        '$select=health,count(tree_id)' +\
        '&$where=boroname=\'Queens\'' +\
        '&$group=health').replace(' ', '%20')
soql_url_qn = pd.read_json(soql_url)
soql_url_qn.dropna()
```

```python
soql_url = ('https://data.cityofnewyork.us/resource/nwxe-4ae8.json?' +\
        '$select=health,count(tree_id)' +\
        '&$where=boroname=\'Manhattan\'' +\
        '&$group=health').replace(' ', '%20')
soql_url_mn = pd.read_json(soql_url)
soql_url_mn.dropna()
```

```python
app = dash.Dash()

colors = {'background':'D3D3D3', 'text':'#000000'}

app.layout = html.Div(children=[

    dcc.Graph(id='Graph',
             figure={
            'data': [
                {'x': soql_url_bx['health'], 'y':soql_url_bx['count_tree_id'], 'type': 'bar', 'name': u'Bronx'},
                {'x': soql_url_bk['health'], 'y':soql_url_bk['count_tree_id'], 'type': 'bar', 'name': u'Brooklyn'},
                {'x': soql_url_mn['health'], 'y':soql_url_mn['count_tree_id'], 'type': 'bar', 'name': u'Manhattan'},
                {'x': soql_url_si['health'], 'y':soql_url_si['count_tree_id'], 'type': 'bar', 'name': u'Staten Island'},
                {'x': soql_url_qn['health'], 'y':soql_url_qn['count_tree_id'], 'type': 'bar', 'name': u'Queens'}
            ],
            'layout': {
                'title': 'NYC Tree Health Condition by Borough',
                'plot_bgcolor':colors['background'],
                'paper_bgcolor':colors['background'],
                'font':{'color':colors['text']}
                
            }
        }
    )
])


if __name__ == '__main__':
    app.run_server()
```
